# Interpolation linéaire

```python
import pandas as pd
import numpy as np

import datetime as dt

import matplotlib.pyplot as plt
import matplotlib as mlp
```

```python
np.random.seed(42)
maturite= np.arange(3, 123, 3)
taux= np.random.random(40)*0.1 + np.linspace(0, 0.2, 40)

df= pd.DataFrame(
{
    "Maturite": maturite
    , "Taux": taux
})
df
```

```python
# extrapolation linéaire
# REVOIR LES COEF C EST PAS BON
df["coef"]= (df["Taux"] - df["Taux"].shift(-1))/(df["Maturite"] - df["Maturite"].shift(-1))
df["c"]= df["Taux"].shift(-1) - df["Maturite"].shift(-1)*df["coef"]
df
```

```python
def interpolation_li(x1, y1, x2, y2, X):
    """
    ATTENTION ICI x1 <= x2 !!!
    return l'output de la fonction ax+b
    a= (df["Taux"] - df["Taux"].shift(-1))/(df["Maturite"] - df["Maturite"].shift(-1))
    b= df["Taux"].shift(-1) - df["Maturite"].shift(-1)*df["coef"]
    """
    a= (y1-y2)/(x1-x2)
    b= y2 - x2*a
    
    return a*X + b
```
